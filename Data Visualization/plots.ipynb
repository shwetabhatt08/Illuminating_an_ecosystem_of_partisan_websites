```python
#demographics plots
import alexa_demographics
from alexa_demographics import *
import numpy as np
import matplotlib.pyplot as plt
import collections
import pandas as pd

def get_bar_charts(left_mean_val, left_std_val, right_mean_val, right_std_val, xticks, title, x_label, ylim):
    left_mean = left_mean_val
    left_std = left_std_val
    N = len(left_mean)
    ind = np.arange(N)  # the x locations for the groups
    width = 0.4    # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, left_mean, width, yerr= left_std)
    right_mean = right_mean_val
    right_std = right_std_val
    rects2 = ax.bar(ind + width, right_mean, width, yerr=right_std)

    # add some text for labels, title and axes ticks
    ax.set_ylabel('Mean demographic score')
    ax.set_xlabel(x_label)
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels(xticks)

    ax.legend((rects1[0], rects2[0]), ('Left', 'Right'))
    plt.ylim(ylim)

    def autolabel(rects):
        """
        Attach a text label above each bar displaying its height
        """
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                    '%d' % int(height),
                    ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)
    #fig.autofmt_xdate(bottom=0.2, rotation=30, ha='right')
    ax.axhline(y=90, color='gray', linestyle='--')
    ax.axhline(y=110, color='gray', linestyle='--')
    #ax.text(1.02, 4.2e4, "42000", va='center', ha="left", bbox=dict(facecolor="w",alpha=0.5),
        #transform=ax.get_yaxis_transform())
    plt.show()
#age_left_mean

def get_ordered_keys(dict1):
    ordered_dict = collections.OrderedDict(sorted(dict1.items()))
    keys = [x for x in ordered_dict]
    return keys

def get_ordered_values(dict1):
    ordered_dict = collections.OrderedDict(sorted(dict1.items()))
    values = []
    for each_item in ordered_dict.items():
        values.append(each_item[1])
    return values
```
